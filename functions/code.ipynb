```python
# functions

def is_even(num):
    "Function to check if a number is even"
    if num % 2 == 0: return 'even'
    else: return 'odd'

is_even(4)    

for i in range(1,1000):
    x=is_even(i)
    print(x)
```

```python
def  multiply(*args):
    "Function to multiply all the numbers passed in argumnets"
    result=1
    for i in args:
        result = result * i

    return result    

multiply(2,3,4,5) 
```

```python
def display(**kwargs):
    "Function to display key value pairs"
    for key, value in kwargs.items():
        print(f"{key} : {value}")



display(name="John", age=30, city="New York")

```
