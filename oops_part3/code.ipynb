```python
# Example for Aggregation in OOPS
class Customer : 

    def __init__(self,name,address,gender):
        self.name=name
        self.address=address
        self.gender=gender

    def print_address(self):
        print('City',self.address.city)
        print('Pin',self.address.pin)
        print('State',self.address.state)


class Address:

    def __init__(self,city,pin,state):
        self.city=city
        self.pin=pin
        self.state=state


add1=Address("Banglore",560068,"Karnataka")
C1=Customer("Deepu",add1,"Male")        
```
