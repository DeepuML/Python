```python
# menu driven program  
fnum =  int(input("Enter first number : "))
snum =  int(input("Enter second number : "))

op = input("Enter the operation : ")

if op == "+" :
    print("This is addition" , fnum + snum )
elif op == "-":
    print("This si subtraction: ", fnum-snum) 
elif op == "*":
    print("This is multiplication :", fnum * snum)

elif op == "/":
    print("This is division : ", fnum / snum)

else : 
    print("Invalid Operation ")    
```

```python
# guessing game 

import random 
jackpot = random.randint(1,100)

guess = int(input("enter you guess"))
counter  = 1

while guess != jackpot : 
    if  guess<jackpot:
        print("Too low")
    else :
        print("Too high")

    guess = int(input("Enter the guess again : "))    
    counter = counter + 1 

else : 
    print("Congratulations! You guessed it right")
    print("You took ", counter , " attempts to guess the number")
```
