```python
# indexing
s = "This is python string"
s[2]
s[1]

for i in range(len(s)):
    print(s[i], sep=' ', end=' ')
```

```python
# #  negative indexing
# s[-1]
# print(s[-1])
# print(s[-2])
print(s)
print("Reversed string:")
for i in range(len(s)-1,-1,-1):
    print(s[i], sep=' ', end=' ')
```

```python
# slicing  
 
s = "This is python string"
print(s[0:4])
print(s[5:7])
print(s[0:7:2])
print(s[:7])
print(s[:])

# negative  slicing
print(s[6:0:-1])
print(s[::-1])
```
