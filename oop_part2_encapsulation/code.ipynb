```python
# A class for coordinate geometry
class Point:

    def __init__(self,x,y):
        self.x=x
        self.y=y

    # <x,y> = Point(2,3)
    def __str__(self):
        return '<{},{}>'.format(self.x,self.y) 
    
    # eculidean distance

    def euclidean_distance(self,other):
        return ((self.x-other.x)**2 + (self.y-other.y)**2)**0.5
    

```

```python
p1=Point(10,10) 
p2=Point(10,111)

print(p1)
print(p2)
print("Distance between p1 and p2 is : ", p1.euclidean_distance(p2))
```
