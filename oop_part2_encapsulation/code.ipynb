```python
# A class for coordinate geometry
class Point:

    def __init__(self,x,y):
        self.x=x
        self.y=y

    # <x,y> = Point(2,3)
    def __str__(self):
        return '<{},{}>'.format(self.x,self.y) 
    
    # eculidean distance
    def euclidean_distance(self,other):
        return ((self.x-other.x)**2 + (self.y-other.y)**2)**0.5
    
    # origin distance 
    def origin_distance(self):
        return (self.x**2 + self.y**2) **0.5

    # Line equation


class Line: 
    #  construcotor

    def __init__(self,A,B,C):
        self.A=A
        self.B=B
        self.C=C

    def __str__(self):
        return '{}x+{}y+{}=0'.format(self.A,self.B,self.C)
    
    def check_equation(self, point):
        equation = self.A*point.x + self.B*point.y + self.C
        if equation == 0: 
            return "Point is on line"
        else :
             return "Point is not on line"
        

```
